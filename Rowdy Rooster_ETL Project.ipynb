Dataset 1

```python
# import file
US = pd.read_csv('US_youtube_trending_data.csv') 
GB = pd.read_csv('GB_youtube_trending_data.csv')
CA = pd.read_csv('CA_youtube_trending_data.csv')

US['country'] = 'US'
GB['country'] = 'GB'
CA['country'] = 'CA'
frames = [US, GB, CA]

#merge
df = pd.concat(frames).drop_duplicates()
```

```python
df.to_csv('US_GB_CA_merged.csv')
```

Dataset 2

load json file

```python
with open("CA_category_id.json")as f:
    data = json.load(f)
```

Get 'id' and 'title' from dictionary

```python
from pprint import pprint
list_of_items = data['items']
```

```python
d = []
for i in list_of_items :
    d.append(
        {
            'id' : i['id'],
            'title': i['snippet']['title']
        }
    )
 
d = pd.DataFrame(d)
```

```python
d
```

```python
d.to_csv('video_category.csv', index=False)
```
